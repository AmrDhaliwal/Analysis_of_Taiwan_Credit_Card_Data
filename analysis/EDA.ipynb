```python
# Created a new dataframe for specific columns I'm interersted in
df1 = df.filter(['AGE', 'MARRIAGE', 'BILL_AMT_APR', 'PAY_AMT_APR', 'PAY_AMT_AUG', 'EDUCATION', 'BILL_AMT_AUG'], axis=1)
```

```python
# Set theme for Seaborn plots
sns.set_theme(style="ticks", palette="pastel")
```

```python
# Draw a nested boxplot to show Marriage by Age.
plt.figure(figsize=(15,10))

sns.boxplot(x="MARRIAGE", y="AGE",
            data=df)
sns.despine(offset=10, trim=True)
```

```python
# Distribution plot to show Age.
plt.figure(figsize=(20,15))

sns.displot(df1['AGE'], kde=False, bins=25)
```

```python
# Replaced all 0's, 5 and 6's with the number 4 since 4 means other according to the
# original dataset and 5 and 6 weren't specified so, I 
df1["EDUCATION"].replace({0: 4, 5: 4, 6: 4}, inplace=True)
```

```python
df1.describe().T
```

```python
# Created a scatter plot to show Age by BILL_AMT_APR.
plt.figure(figsize=(20,15))

sns.scatterplot(data=df1, x="AGE", y="BILL_AMT_APR", hue='MARRIAGE')
```

```python
# Created a scatter plot to show Age by PAY_AMT_APR.
plt.figure(figsize=(20,15))

sns.scatterplot(data=df1, x="AGE", y="PAY_AMT_APR", hue='MARRIAGE')
```
