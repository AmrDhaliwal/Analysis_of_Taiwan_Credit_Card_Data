```python
# Set theme for Seaborn plots
sns.set_theme(style='ticks', palette='pastel')
```

```python
# Draw a nested boxplot to show Marriage by Age.
plt.figure(figsize=(15,10))

ax = sns.boxplot(x='MARRIAGE', y='AGE',
            data=df)
sns.despine(offset=10, trim=True)
ax.set(ylabel='AGE in Years')
plt.show()
```

```python
# Distribution plot to show Age.
plt.figure(figsize=(20,15))

bx = sns.displot(df['AGE'], kde=False, bins=25)
bx.set(xlabel='AGE in Years')
plt.show()
```

```python
# Replaced all 0's, 5 and 6's with the number 4 since 4 means other according to the
# original dataset and 5 and 6 weren't specified so, I 
df['EDUCATION'].replace({0: 4, 5: 4, 6: 4}, inplace=True)
```

```python
df.describe().T
```

```python
# Created a scatter plot to show Age by BILL_AMT_APR.
plt.figure(figsize=(20,15))

cx = sns.scatterplot(data=df, x='AGE', y='BILL_AMT_APR', hue='MARRIAGE')
cx.set(xlabel='AGE in Years', ylabel='Bill_Amt_Apr in NT$')
plt.show()
```

```python
# Created a scatter plot to show Age by PAY_AMT_APR.
plt.figure(figsize=(20,15))

dx = sns.scatterplot(data=df, x='AGE', y='PAY_AMT_APR', hue='MARRIAGE')
dx.set(xlabel='AGE in Years', ylabel='Pay_Amt_Apr in NT$')
plt.show()
```

```python
# Created a box plot to show EDUCATION by BILL_AMT_APR.
plt.figure(figsize=(15,10))

ex = sns.boxplot(x='EDUCATION', y='BILL_AMT_APR',
            data=df)
sns.despine(offset=10, trim=True)
ex.set(ylabel='Bill_Amt_Apr in NT$')
plt.show()
```

```python
# Created a scatter plot to show EDUCATION by PAY_AMT_APR with an AGE hue.
plt.figure(figsize=(20,15))

fx = sns.scatterplot(data=df, x='EDUCATION', y='PAY_AMT_APR', hue='AGE')
fx.set(ylabel='Pay_Amt_Apr in NT$')
plt.show()
```

```python
# Created a scatter plot to show EDUCATION by PAY_AMT_AUG with an AGE hue.
plt.figure(figsize=(20,15))

gx = sns.scatterplot(data=df, x='EDUCATION', y='PAY_AMT_AUG', hue='AGE')
gx.set(ylabel='Pay_Amt_Aug in NT$')
plt.show()
```

```python
# Created a scatter plot to show EDUCATION by BILL_AMT_APR with an AGE hue.
plt.figure(figsize=(20,15))

hx = sns.scatterplot(data=df, x='EDUCATION', y='BILL_AMT_APR', hue='AGE')
hx.set(ylabel='Bill_Amt_Apr in NT$')
plt.show()
```

```python
# Created a scatter plot to show EDUCATION by BILL_AMT_AUG with an AGE hue.
plt.figure(figsize=(20,15))

ix = sns.scatterplot(data=df, x='EDUCATION', y='BILL_AMT_AUG', hue='AGE')
ix.set(ylabel='Bill_Amt_Aug in NT$')
plt.show()
```

```python
# Histplot to show the credit card limits
plt.figure(figsize = (15,10))
plt.title('Credit Limit')
jx = sns.histplot(df['Limit_BAL'],kde=True,bins=200, color='orange')
jx.set(xlabel='Limit_Bal in NT$')
plt.show()


```

```python
# A boxplot distribution to show AGE vs credit card limit
plt.figure(figsize = (15,10))
kx = sns.boxplot(x='SEX', y='Limit_BAL', hue='SEX',data=df,showfliers=True)
kx.set(xlabel='SEX by Gender', ylabel='Limit_Bal by NT$')
plt.show()
```

```python
# A boxplot showing correlation between AGE and Limit_BAL with SEX as the hue.
plt.figure(figsize=(20,15))

lx = sns.boxplot(data=df, x='AGE', y='Limit_BAL', hue='SEX')
lx.set(xlabel='AGE in Years', ylabel='Limit_Bal in NT$')
plt.show()
```
